```python
import pandas as pd
from collections import Counter
```

```python
all_markers = pd.read_csv('GSE156337.Reduced.All_Markers.csv',index_col=0)
all_markers = all_markers[all_markers['avg_log2FC'] >0 ]
all_markers
```

```python
gene_occurrence = dict(Counter(all_markers['gene']))
unique_markers = [gene for gene in gene_occurrence if gene_occurrence[gene] == 1]
```

```python
len(unique_markers)
```

```python
cd4 = all_markers[all_markers['cluster'] == 'CD4+ Cells']
cd4 = cd4.sort_values('avg_log2FC',ascending=False)
cd4
```

```python
cd8 = all_markers[all_markers['cluster'] == 'CD8+ Cells']
cd8 = cd8.sort_values('avg_log2FC',ascending=False)
cd8
```

```python
set(cd4['gene'][:100]).intersection(set(cd8['gene'][:100]))
```
