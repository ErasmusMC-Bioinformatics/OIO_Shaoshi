# ScRNA-seq H5AD

This notebook shows how to pack the expression matrix of scRNA-seq into H5AD files. This is not only for convenience, but also for the efficiency of memory usage if the expression matrix is big.

This notebook illustrates the preprossessing of Liver Cell Atlas (GSE115469 in GEO database).

Please download the expression matrix in CSV files:
https://ftp.ncbi.nlm.nih.gov/geo/series/GSE115nnn/GSE115469/suppl/GSE115469_Data.csv.gz

```python
import scanpy as sc
import pandas as pd
```

```python
adata = sc.read_csv('GSE115469_Data.csv.gz').T
```

```python
annotation = pd.read_csv('GSE115469_Annoation.csv',index_col=0)
```

```python
adata.obs = annotation
```
