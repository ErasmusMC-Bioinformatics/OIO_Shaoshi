```python
import pandas as pd
import gseapy as gp
import os
```

```python
signature = pd.read_csv('GSE146409.Immune_Signature.txt',delimiter='\t',index_col=0)
signature
```

```python
cell_types = list(signature.columns.values)
cell_types
```

```python
library = 'WikiPathway_2021_Human'
```

```python
try:
    os.mkdir(library)
except:
    pass

for cell_type in cell_types:
    signature_matrix = signature.filter([cell_type])
    signature_matrix = signature_matrix[signature_matrix[cell_type] != 1]
    signature_gene_list = list(signature_matrix.index.values)
    enr = gp.enrichr(gene_list=signature_gene_list, gene_sets=library,
                 organism='Human', description='test_name',
                 outdir='%s/%s'%(library,cell_type),
                 # no_plot=True,
                 cutoff=0.5 # test dataset, use lower value from range(0,1)
                )
```
